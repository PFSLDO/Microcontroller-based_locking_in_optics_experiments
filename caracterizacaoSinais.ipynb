# Caracterizacao dos Sinais 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataHorizontal = pd.read_csv('./1110_H.csv', sep = ",")
dataVertical = pd.read_csv('./1110_V.csv', sep = ",")

display(dataHorizontal)
display(dataVertical)
```

```python
tempo = dataHorizontal['-247.19999731E-3']
sinalAzulHorizontal = dataHorizontal['1.9399999']
sinalAmareloHorizontal = dataHorizontal['23.00000000']

# Plotar o sinal original
plt.figure(figsize=(12, 5))
plt.plot(tempo, sinalAzulHorizontal, label='Sinal de saída', color = 'b')
plt.plot(tempo, sinalAmareloHorizontal, label='Sinal de entrada', color = 'y')
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.title('Sinal de Tensão - Polarizacao horizontal')
plt.legend()
plt.grid()
plt.show()

# Calcular a Transformada de Fourier
frequencia_amostragemH = 1 / (tempo[1] - tempo[0])  # assumindo espaçamento constante
fft_sinalAzulH = np.fft.fft(sinalAzulHorizontal)
frequenciasAzulH = np.fft.fftfreq(len(sinalAzulHorizontal), d=1/frequencia_amostragemH)
fft_sinalAmareloH = np.fft.fft(sinalAmareloHorizontal)
frequenciasAmareloH = np.fft.fftfreq(len(sinalAmareloHorizontal), d=1/frequencia_amostragemH)

# Plotar a magnitude da Transformada de Fourier
plt.figure(figsize=(12, 5))
plt.plot(frequenciasAzulH, np.abs(fft_sinalAzulH), label='Transformada de Fourier azul Horizontal', color = 'b')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.title('Espectro de Frequência')
plt.legend()
plt.grid()
plt.show()
plt.figure(figsize=(12, 5))
plt.plot(frequenciasAmareloH, np.abs(fft_sinalAmareloH), label='Transformada de Fourier amarelo Horizontal', color = 'y')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.title('Espectro de Frequência')
plt.legend()
plt.grid()
plt.show()
```

```python
tempo = dataVertical['-122.2000048E-3']
sinalAzulVertical = dataVertical['1.3600000']
sinalAmareloVertical = dataVertical['26.80000114']

# Plotar o sinal original
plt.figure(figsize=(12, 5))
plt.plot(tempo, sinalAzulVertical, label='Sinal de saída', color = 'b')
plt.plot(tempo, sinalAmareloVertical, label='Sinal de entrada', color = 'y')
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.title('Sinal de Tensão - Polarizacao vertical')
plt.legend()
plt.grid()
plt.show()

# Calcular a Transformada de Fourier
frequencia_amostragemV = 1 / (tempo[1] - tempo[0])  # assumindo espaçamento constante
fft_sinalAzulV = np.fft.fft(sinalAzulVertical)
frequenciasAzulV = np.fft.fftfreq(len(sinalAzulVertical), d=1/frequencia_amostragemH)
fft_sinalAmareloV = np.fft.fft(sinalAmareloVertical)
frequenciasAmareloV = np.fft.fftfreq(len(sinalAmareloVertical), d=1/frequencia_amostragemH)

# Plotar a magnitude da Transformada de Fourier
plt.figure(figsize=(12, 5))
plt.plot(frequenciasAzulV, np.abs(fft_sinalAzulV), label='Transformada de Fourier azul Vertical', color = 'b')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.title('Espectro de Frequência')
plt.legend()
plt.grid()
plt.show()
plt.figure(figsize=(12, 5))
plt.plot(frequenciasAmareloV, np.abs(fft_sinalAmareloV), label='Transformada de Fourier amarelo Vertical', color = 'y')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.title('Espectro de Frequência')
plt.legend()
plt.grid()
plt.show()
```

```python
#Tratando o sinal antes de fazer a transformada de Fourier para dar enfoque nas frequências longe do 0

#Método 1: retirar a média do sinal antes de fazer a transformada
```

```python
sinal_azul_sem_dcH = sinalAzulHorizontal - np.mean(sinalAzulHorizontal)
fft_sinalAzul_sem_dcH = np.fft.fft(sinal_azul_sem_dcH)
frequenciasAzulSemDCH = np.fft.fftfreq(len(sinal_azul_sem_dcH), d=1/frequencia_amostragemH)

sinal_amarelo_sem_dcH = sinalAmareloHorizontal - np.mean(sinalAmareloHorizontal)
fft_sinalAmarelo_sem_dcH = np.fft.fft(sinal_amarelo_sem_dcH)
frequenciasAmareloSemDCH = np.fft.fftfreq(len(sinal_amarelo_sem_dcH), d=1/frequencia_amostragemH)

# Plotar a magnitude da Transformada de Fourier
plt.figure(figsize=(12, 5))
plt.plot(frequenciasAzulSemDCH, np.abs(fft_sinalAzul_sem_dcH), label='Transformada de Fourier azul Horizontal - sem a média', color = 'b')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.title('Espectro de Frequência')
plt.legend()
plt.grid()
plt.show()
plt.figure(figsize=(12, 5))
plt.plot(frequenciasAmareloSemDCH, np.abs(fft_sinalAmarelo_sem_dcH), label='Transformada de Fourier amarelo Horizontal - sem a média', color = 'y')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.title('Espectro de Frequência')
plt.legend()
plt.grid()
plt.show()
```

```python
sinal_azul_sem_dcV = sinalAzulVertical - np.mean(sinalAzulVertical)
fft_sinalAzul_sem_dcV = np.fft.fft(sinal_azul_sem_dcV)
frequenciasAzulSemDCV = np.fft.fftfreq(len(sinal_azul_sem_dcV), d=1/frequencia_amostragemV)

sinal_amarelo_sem_dcV = sinalAmareloVertical - np.mean(sinalAmareloVertical)
fft_sinalAmarelo_sem_dcV = np.fft.fft(sinal_amarelo_sem_dcV)
frequenciasAmareloSemDCV = np.fft.fftfreq(len(sinal_amarelo_sem_dcV), d=1/frequencia_amostragemV)

# Plotar a magnitude da Transformada de Fourier
plt.figure(figsize=(12, 5))
plt.plot(frequenciasAzulSemDCV, np.abs(fft_sinalAzul_sem_dcV), label='Transformada de Fourier azul Vertical - sem a média', color = 'b')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.title('Espectro de Frequência')
plt.legend()
plt.grid()
plt.show()
plt.figure(figsize=(12, 5))
plt.plot(frequenciasAmareloSemDCV, np.abs(fft_sinalAmarelo_sem_dcV), label='Transformada de Fourier amarelo Vertical - sem a média', color = 'y')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.title('Espectro de Frequência')
plt.legend()
plt.grid()
plt.show()
```
