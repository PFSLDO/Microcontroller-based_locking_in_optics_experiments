# Caracterizacao dos Sinais 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataHorizontal = pd.read_csv('./1110_H.csv', sep = ",")
dataVertical = pd.read_csv('./1110_V.csv', sep = ",")

display(dataHorizontal)
display(dataVertical)
```

```python
tempo = dataHorizontal['-247.19999731E-3']
sinalAzul = dataHorizontal['1.9399999']
sinalAmarelo = dataHorizontal['23.00000000']

# Plotar o sinal original
plt.figure(figsize=(12, 5))
plt.plot(tempo, sinalAzul, label='Sinal de saÃ­da', color = 'b')
plt.plot(tempo, sinalAmarelo, label='Sinal de entrada', color = 'y')
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.title('Sinal de TensÃ£o')
plt.legend()
plt.grid()
plt.show()

# Calcular a Transformada de Fourier
frequencia_amostragem = 1 / (tempo[1] - tempo[0])  # assumindo espaÃ§amento constante
fft_sinalAzul = np.fft.fft(sinalAzul)
frequenciasAzul = np.fft.fftfreq(len(sinalAzul), d=1/frequencia_amostragem)
fft_sinalAmarelo = np.fft.fft(sinalAmarelo)
frequenciasAmarelo = np.fft.fftfreq(len(sinalAmarelo), d=1/frequencia_amostragem)

# Plotar a magnitude da Transformada de Fourier
plt.figure(figsize=(12, 5))
plt.plot(frequenciasAzul, np.abs(fft_sinalAzul), label='Transformada de Fourier azul', color = 'b')
plt.xlabel('FrequÃªncia (Hz)')
plt.ylabel('Magnitude')
plt.title('Espectro de FrequÃªncia')
plt.legend()
plt.grid()
plt.show()
plt.figure(figsize=(12, 5))
plt.plot(frequenciasAmarelo, np.abs(fft_sinalAmarelo), label='Transformada de Fourier amarelo', color = 'y')
plt.xlabel('FrequÃªncia (Hz)')
plt.ylabel('Magnitude')
plt.title('Espectro de FrequÃªncia')
plt.legend()
plt.grid()
plt.show()
```
